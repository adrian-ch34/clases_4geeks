- 1. Cuál es la probabilidad de obetener un número impar en un lanzamiento de dados ?

```python
s = {1,2,3,4,5,6} #Espacio muestra 
e = {1,3,5} #Evento

#La cardinalidad de la interseccion del evento y el espacio muestra, entre la cardinalidad del espacio muestra
Fraction(len(set.intersection(e,s)),len(s))
```

- 2. Cuál es la probabilidad de obetener un número primo o un número par en un lanzamiento de dados ?

```python
s = {1,2,3,4,5,6} #Espacio muestra 
e_1 = {2,3,5} #Evento
e_2 = {2,4,6} #Evento
e = set.union(e_1,e_2)
c_n = len(set.intersection(e,s))#La cardinalidad de la interseccion del evento y el espacio muestra, entre la cardinalidad del espacio muestra
c_d = len(s)

Fraction(c_n,c_d)
```

- 3. Cuál es la probabilidad de obetener un número par y primo en un lanzamiento de dado?

```python
s = {1,2,3,4,5,6}
e_1 = {2,4,6}
e_2 = {2,3,5}
e = set.intersection(e_1,e_2) #Cuando es "Y" es interseccion y cuando es "O" es union
c_n = len(set.intersection(e,s)) #La cardinalidad de la interseccion del evento y el espacio muestra, entre la cardinalidad del espacio muestra
c_d = len(s)

Fraction(c_n,c_d)
```

```python
def Pr(event:set, space:set) -> Fraction:
    return Fraction(len(set.intersection(event,space)), len(space))
```

```python
Pr({2,4,6},{1,2,3,4,5,6})
```

- 4. El método de Montecarlo nos permite aproximar una probabilidad en base a la repetición continua de un evento. Cuál es la probabilidad de obtener un número par en un lanzamiento de dados ? 

```python
event = [2,4,6]
space = [1,2,3,4,5,6]
frec = []
trials = list(range(1,1001))
for trial in trials:
    total = 0
    for i in range(trial):
        if random.choice(space) in event:
            total = total + 1
    frec.append(total/trial)
```

```python
plt.plot(trials,frec)
plt.xlabel("Trials")
plt.ylabel("Relative Frequence")
plt.axhline(y=0.5, color="tomato")
plt.grid(True)
plt.show()
```
