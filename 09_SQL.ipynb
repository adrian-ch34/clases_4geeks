```python
from pandasql import sqldf
import os
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/syntagmatic/gapminder-csv/refs/heads/master/gapminder.csv"
df = pd.read_csv(url)
df.info()
```

sqldf() es la función principal para poder correr en entornos python comandos SQL, este nos permite navegar en base a querys un conjunto de datos. El objetivo de implementar SQL a nuestro trabajo de análisis es exportar los datos que necesitamos en nuestras consultas y poder trabajarlas con pandas para la visualización y estadística.  

```python
sqldf("SELECT * FROM df LIMIT 5")
```

```python
#Trabajamos con las columnas:
# children_per_woman_total_fertility, life_expectancy_years

sqldf("SELECT children_per_woman_total_fertility, life_expectancy_years FROM df LIMIT 10")
```

```python
sqldf("SELECT geo, time, children_per_woman_total_fertility, life_expectancy_years FROM df LIMIT 10")
```

sqldf("SELECT (cols) FROM (table) WHERE (filtros)")

```python
sqldf("SELECT geo, time, children_per_woman_total_fertility, life_expectancy_years FROM df WHERE geo='arg' AND time > 1795 LIMIT 10")
```

```python
sqldf("SELECT geo,time,children_per_woman_total_fertility, life_expectancy_years FROM df WHERE geo IN ('col','arg') AND time = 2000")
```

```python
sqldf("SELECT geo,time,children_per_woman_total_fertility, life_expectancy_years FROM df WHERE geo IN ('col','arg','mex') AND time = 2000")
```

```python
sqldf("SELECT DISTINCT geo FROM df")
```

```python
sqldf("SELECT geo,time,gdp_total_yearly_growth FROM df WHERE geo='per' AND time BETWEEN 1985 AND 2000 ORDER BY time DESC")
```

```python
sqldf("SELECT geo,time,gdp_total_yearly_growth FROM df WHERE geo='arg' AND time BETWEEN 1990 AND 2010 ORDER BY time DESC")
```

```python
sqldf("SELECT MAX (gdp_total_yearly_growth) AS max_gdp FROM df WHERE geo='mex' ")
```

```python
sqldf("SELECT MAX (gdp_total_yearly_growth) AS max_gdp, geo FROM df WHERE geo='mex' ")
```

```python
sqldf("SELECT MAX (gdp_total_yearly_growth) AS gdp_max, geo, time FROM df WHERE geo IN ('arg','mex','per') AND time >= 1985 LIMIT 20")
```

```python
sqldf("SELECT MAX (gdp_total_yearly_growth) AS gdp_max, geo, time FROM df WHERE geo IN ('arg','mex','per') AND time >= 1985 GROUP BY geo")
```

```python
my_df = sqldf("SELECT geo, time, children_per_woman_total_fertility, life_expectancy_years, gdp_total_yearly_growth FROM df WHERE geo IN ('JPN','bra','usa') AND time > 1975 AND life_expectancy_years IS NOT NULL AND gdp_total_yearly_growth IS NOT NULL ORDER BY geo ")
```
